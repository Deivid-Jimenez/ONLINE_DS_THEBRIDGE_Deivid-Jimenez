```python
class Tienda:
 
    Tipo = "Electrodomésticos"
    Abierta = True

    def __init__(self, nombre, direccion, num_empleados, ventas_ultimos_3_meses):
        # Atributos propios de cada tienda
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_ultimos_3_meses = ventas_ultimos_3_meses

    def calcular_ventas_totales(self):
        # Método para calcular las ventas totales de los últimos 3 meses
        return sum(self.ventas_ultimos_3_meses)

    def calcular_media_ventas_por_empleado(self):
        # Método para calcular la media de ventas por empleado
        ventas_totales = self.calcular_ventas_totales()
        return ventas_totales / self.num_empleados

    def info_tienda(self):
        # Método que devuelve el nombre y la dirección de la tienda
        return f"Nombre: {self.nombre}, Dirección: {self.direccion}"

    def ventas_ultimo_mes(self):
        # Método que obtiene las ventas del último mes usando un bucle for
        ultima_venta = None
        for venta in self.ventas_ultimos_3_meses:
            ultima_venta = venta
        return ultima_venta

    def proyeccion_ventas_marketing(self, x):
        # Método para calcular la proyección de ventas con inversión en marketing
        factor = 1.2 if x < 1000 else 1.5
        self.ventas_ultimos_3_meses = [venta * factor for venta in self.ventas_ultimos_3_meses]
        return self.ventas_ultimos_3_meses

# Ejemplo de uso
tienda1 = Tienda("ElectroCity", "Calle Castelló 123", 10, [20000, 25000, 30000])
tienda2 = Tienda("ElectroNorte", "Calle San Jacobo 123", 12, [60000, 35000, 40000])
tienda2 = Tienda("ElectroSur", "Calle Jorge Juan 123", 12, [80000, 35000, 40000])
print(tienda1.info_tienda()) 
print(f"Ventas del último mes: {tienda1.ventas_ultimo_mes()}")  
inversion_marketing = 1200
ventas_proyectadas = tienda1.proyeccion_ventas_marketing(inversion_marketing)
print(f"Ventas proyectadas: {ventas_proyectadas}")
print(tienda2.info_tienda()) 
print(f"Ventas del último mes: {tienda2.ventas_ultimo_mes()}")  
inversion_marketing = 1200
ventas_proyectadas = tienda2.proyeccion_ventas_marketing(inversion_marketing)
print(f"Ventas proyectadas: {ventas_proyectadas}")
print(tienda3.info_tienda()) 
print(f"Ventas del último mes: {tienda3.ventas_ultimo_mes()}")  
inversion_marketing = 1200
ventas_proyectadas = tienda3.proyeccion_ventas_marketing(inversion_marketing)
print(f"Ventas proyectadas: {ventas_proyectadas}")
```

Ejercicio 2
En este ejercicio vamos a implementar una clase Perro en Python. La clase tiene las siguientes características:

Cosas que sabemos seguro que tiene un perro

* Tiene 4 patas
* 2 orejas
* 2 ojos
* Una velocidad de 0. Por defecto, el perro está parado
+ Cuando se inicialice:

+ El perro será de una determinada raza
Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
Por defecto no tendrá dueño, a no ser que se diga lo contrario.
Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.

*Necesita otro método (parar), donde pondremos la velocidad a 0.

*Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.

Se pide:

*I mplementa la clase Perro
* Crea un objeto de tipo Perro, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo


```python
class Perro:
    # Atributos comunes
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0

    def __init__(self, raza, color_pelo="Marrón", dueño=None):
        self.raza = raza
        self.color_pelo = color_pelo
        self.dueño = dueño

    def andar(self, aumento_velocidad):
        self.velocidad += aumento_velocidad

    def parar(self):
        self.velocidad = 0

    def ladrar(self, sonido):
        return f"GUAUI {sonido}"

# Crear un objeto de tipo Perro sin dueño
mi_perro = Perro(raza="Labrador")

# Comprobar los atributos del perro
print(f"Raza: {mi_perro.raza}")  # Raza: Labrador
print(f"Color del pelo: {mi_perro.color_pelo}")  # Color del pelo: Marrón
print(f"Dueño: {mi_perro.dueño}")  # Dueño: None
print(f"Velocidad inicial: {mi_perro.velocidad}")  # Velocidad inicial: 0

# Probar el método andar
mi_perro.andar(10)
print(f"Velocidad después de andar: {mi_perro.velocidad}")  # Velocidad después de andar: 10

# Probar el método parar
mi_perro.parar()
print(f"Velocidad después de parar: {mi_perro.velocidad}")  # Velocidad después de parar: 0

# Probar el método ladrar
print(mi_perro.ladrar("¡Guau!"))  # GUAUI ¡Guau!
```
