```python
datos_faltantes = df_viajes_aereos.isnull().sum()
print("Datos faltantes por columna:")
print(datos_faltantes)
```

```python
# Calcular el porcentaje de datos faltantes por columna
porcentaje_faltantes = (df_viajes_aereos.isnull().sum() / len(df_viajes_aereos)) * 100
print("\nPorcentaje de datos faltantes por columna:")
print(porcentaje_faltantes)
```

```python
# Explorar columnas con posibles problemas
for columna in df_viajes_aereos.columns:
    print(f"\nValores únicos en la columna '{columna}':")
    print(df_viajes_aereos[columna].value_counts(dropna=False).head(10))  # Mostrar los 10 valores más frecuentes
```

Vamos con los nulos o NaN localizados en 1.2. Elimina las filas que tengan más de dos nulos, las podemos considerar demasiado ruidosas y las quitamos.

```python

df_limpio = df_sin_duplicados.dropna(thresh=len(df_sin_duplicados.columns)-2)


print(f"Número de filas después de eliminar filas con más de dos nulos: {len(df_limpio)}")


print(df_limpio.head())
```

Ahora con los nulos en las columnas numéricas, ¿hay alguna columna que creas que puedes reconstruir completamente? ¿Cuál?

```python
media_distancia_por_ruta = df_limpio.groupby(['Origen', 'Destino'])['Distancia'].transform('mean')


df_limpio.loc[:, 'Distancia'] = df_limpio['Distancia'].fillna(media_distancia_por_ruta)


print("Valores nulos en 'Distancia' después de la imputación:")
print(df_limpio['Distancia'].isnull().sum())
```

```python
df_limpio['consumo_kg'] = (
    df_limpio['consumo_kg']
    .astype(str)
    .str.replace(',', '.') 
    .replace('', np.nan) 
    .astype(float)
)

# Verificar los valores únicos en 'consumo_kg'
print("Valores únicos en 'consumo_kg' después de la limpieza:")
print(df_limpio['consumo_kg'].unique())
```

```python
df_limpio['duracion'] = (
    df_limpio['duracion']
    .astype(str) 
    .str.replace('[^0-9.]', '', regex=True) 
    .replace('', np.nan)
    .astype(float) 
)

# Verificar los valores únicos en 'duracion'
print("Valores únicos en 'duracion' después de la limpieza:")
print(df_limpio['duracion'].unique())
```

```python
df_limpio['avion'] = (
    df_limpio['avion']
    .astype(str) 
    .str.strip() 
    .str.upper()  
)
print("Valores únicos en 'avion' después de la limpieza:")
print(df_limpio['avion'].unique())
```

```python
df_limpio['Aircompany'] = (
    df_limpio['Aircompany']
    .astype(str)
    .str.strip()
    .str.upper() 
)
print("Valores únicos en 'Aircompany' después de la limpieza:")
print(df_limpio['Aircompany'].unique())
```
