## Manipulación de Valores

Para terminar la sesión, veamos cómo usar los filtros para cambiar los valores del **dataframe** de una forma selectiva, lo que nos será muy útil en el futuro para hacer "limpieza".





Supongamos que nos avisan de que los datos de TabarAir son erróneos porque para los vuelos a Los Ángeles en realidad son a San Francisco. ¿Cómo podríamos cambiar ese valor?

```python
es_TabarAir=df_aviones.Aircompany=="TabarAir"
a_LosAngeles=df_aviones.Destino=="Los Angeles"
df_aviones.loc[es_TabarAir & a_LosAngeles,"Destino"]="San Francisco"
df_aviones.loc[es_TabarAir & a_LosAngeles]
df_aviones.head(20)
```

Y ahora nos dicen que creemos una columna nueva "SuperTrayectos" que debe ser True cuando el trayecto es mayor a 12000km y False en caso contrario

```python
## Creandolo con un filtro/seleccion

df_aviones["SuperTrayectos"]= False
df_aviones.loc[df_aviones.Distancia>12000,"SuperTrayectos"]=True
df_aviones
```

```python
df_aviones.loc[df_aviones.SuperTrayectos,"Destino"].value_counts()
```

```python
#Version Alternativa

df_aviones["Supertrayectos"]=df_aviones.Distancia>12000
df_aviones
```
