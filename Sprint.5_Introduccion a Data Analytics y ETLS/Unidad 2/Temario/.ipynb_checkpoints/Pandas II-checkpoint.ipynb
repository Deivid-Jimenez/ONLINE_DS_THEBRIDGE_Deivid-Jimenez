## Columnas

Como en casi todos los notebooks de esta unidad, jugaremos con el dataset de aviones. Ejecuta la siguiente celda:

Añadir una columna nueva que tenga el mismo valor para todas

```python
df_aviones.head()
```

```python
df_aviones["En_hora"] = True
df_aviones
```

Ahora calculemos el consumo * kilometro

```python
df_aviones["consumo_kg"] = pd.to_numeric(df_aviones["consumo_kg"], errors='coerce')
df_aviones["Distancia"] = pd.to_numeric(df_aviones["Distancia"], errors='coerce')
df_aviones["Consumo_por_kilometro"] = df_aviones ["consumo_kg"]/df_aviones["Distancia"]
df_aviones = df_aviones.dropna(subset=["consumo_kg", "Distancia"])
df_aviones
```

Si queremos cambiar el nombre de las columnas

```python
df_aviones.rename(columns={"Consumo_por_kilometro":"Consumo_medio","avion":"Avion","consumo_kg":"Consumo_kg","duracion":"Duracion"},inplace=True)
```

Pero esto solo lo cambia en dicha celda, no modifica la tabla como tal, para ello usamos el argumento inplace

```python
### Reasignando

df_aviones=df_aviones.rename(columns={"Consumo_por_kilometro":"Consumo_medio","avion":"Avion","consumo_kg":"Consumo_kg","duracion":"Duracion"})
```

### Operaciones sencillas

```python
### Cual es la mayor distancia recorrida
df_aviones["Distancia"].max()
```

```python
### Cual es el menor consumo

df_aviones["Consumo_kg"].min()
```

```python
###Cuanta distancia se ha recorrido en los 1200 vuelos

df_aviones["Distancia"].sum()
```

```python
# Cual es la media recorrida por estos viajes
df_aviones["Distancia"].mean()
```

```python
#Consumo Medio
df_aviones["Consumo_kg"].mean()
```

Bueno, como medidas agregadas están bien, pero si quiero algo más a detalle podemos hacer lo siguiente

```python
### Cual es el viaje con menor consumo

viaje = df_aviones["Consumo_kg"].idxmin()
viaje 
```

```python
df_aviones.loc[viaje]
```

```python
## Cual es el viaje con mayor consumo medio

viaje = df_aviones["Consumo_medio"].idxmax()
viaje 

df_aviones.loc[viaje]
```

Pero si solo queremos un valor

```python
df_aviones.loc[viaje,"Avion"].unique()
```

```python
valor_avion = df_aviones.loc[viaje,"Avion"].unique()[0]
print(valor_avion)
```
