## 2 Texto
Vamos a leer ahora el fichero indicado en la ruta contenida en fichero_texto, el objetivo es procesar el contenido para construir un DataFrame con las siguientes columnas direcciones de correo del emisor, veredicto de si es spam o no y confidencia en ese veredicto.

1. Crea un diccionario con las siguientes claves: "from","to","is_spam","confidence" y que cada una tenga una lista vacia.

```python
diccionario_spam = {
    "from": [],
    "to": [],
    "is_spam": [],
    "confidence": []
}
```

2. Lee el fichero (usando with open...) y muestra las primeras 30 filas por pantalla, así podrás ver como es más o menos la estructura del mismo

```python
with open(fichero_texto) as f:
    indice = 0
    for linea in f:
        print(linea, end = "")
        indice += 1
        if indice > 60:
            break
```

Ahora sabemos que el emisor viene después de "To:" el receptor después de "From:" el resultado de Spam tras "X-DSPAM-Result" y el nivel de confianza tras "X-DSPAM-Confidence", crea un programa que lea el fichero y a medida que va leyendo el texto añada cada valor (emisor, receptor, si es spam o no, nivel de confianza) a cada una de las listas del diccionario creado en 1.
Por ejemplo, cuando llegue a una línea con "To:" que añada el correo a ["to"]

```python
import re
claves = {
    "From:": "from",
    "To:": "to",
    "X-DSPAM-Confidence:": "confidence",
    "X-DSPAM-Result:": "is_spam"
}
with open(fichero_texto) as f:
    for linea in f:
        parser = re.match("(From:|To:|X-DSPAM-Result:|X-DSPAM-Confidence:)(.+)", linea)
        if parser:
            diccionario_spam[claves[parser.group(1)]].append(parser.group(2)) if "Result" not in parser.group(1) else diccionario_spam[claves[parser.group(1)]].append(parser.group(2) == "Inocent")
```

4. Recorre el diccionario y muestra el número de valores de cada lista (no los valores, ojo). Es para que compruebes que todos tienen la misma longitud

```python
for clave,valor in diccionario_spam.items():
    print(f"clave: {clave}, tienen {len(valor)} elementos")
```

5. 
Recorre de nuevo los eventos y crea un dataframe con las columnas de las claves del attrib de un "Event" y como valores los valores de los atributos de los 1672 eventos que hay colgando del nodo raíz.

```python
nodo
```

```python
eventos_data ={}
num_eventos = 0
for evento in nodo:
    
    col_to_fill = list(eventos_data.keys())
    for campo,valor in evento.attrib.items(): # recorro los eventos
        if campo not in eventos_data: 
            # la primera vez que aparece el campo tiene en cuenta si hay eventos anteriores sin ese campo
            eventos_data[campo] = [""] * num_eventos + [valor]
            print(campo, eventos_data[campo])
        else:
            # el campo ya ha aparecido antes
            if len(eventos_data[campo]) < num_eventos: # si hay eventos anteriores sin ese campo
                eventos_data[campo] += [""] * (num_eventos - len(eventos_data[campo])) + [valor]
            else:
                eventos_data[campo].append(valor) # los eventos anteriores tienen ese campo
        if campo in col_to_fill: # cuando acabe este for hemos tomado todos los campos del evento
            col_to_fill.remove(campo) 
    for col in col_to_fill: # pero puede haber campos de eventos anteriores que este evento no tenga
        eventos_data[col].append("") 
    num_eventos += 1
```
