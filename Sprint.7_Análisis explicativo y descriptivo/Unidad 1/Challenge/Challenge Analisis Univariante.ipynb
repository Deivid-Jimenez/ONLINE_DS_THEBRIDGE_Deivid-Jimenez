```python
# Calcular cantidad y porcentaje de nulos
nulos_age = df_titanic['age'].isnull().sum()
nulos_age_percent = (nulos_age / len(df_titanic)) * 100

print(f"Valores nulos en 'age': {nulos_age} ({nulos_age_percent:.2f}%)")

# Calcular cantidad y porcentaje de nulos
nulos_deck = df_titanic['deck'].isnull().sum()
nulos_deck_percent = (nulos_deck / len(df_titanic)) * 100

print(f"Valores nulos en 'deck': {nulos_deck} ({nulos_deck_percent:.2f}%)")
```

```python
### En el caso de la columna edad, reemplazamos los nulos por la media

media_age = int(df_titanic['age'].mean())
df_titanic['age'] = df_titanic['age'].fillna(media_age)

# Verificar que ya no hay nulos
print(f"Nulos después de reemplazar: {df_titanic['age'].isnull().sum()}")

### En el caso de deck, eliminamos la columna ya que tenemos un alto porcentaje de nulos
```
