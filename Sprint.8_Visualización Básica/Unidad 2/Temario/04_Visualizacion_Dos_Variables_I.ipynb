### Catplot comparativa de frecuencias con Seaborn

Antes de empezar con las grficas no est de ms recordar como obtener las frecuencias de una variable (las de dos lo veremos en las tablas de contingencia):


```python
# Obtener las freceuncias absolutas y relativas de la variable/columna "state" del dataset df_seguros

frecuencias_absolutas = df_seguros["state"].value_counts()
frecuencias_relativas = df_seguros["state"].value_counts(normalize = True) *  100

print(frecuencias_absolutas)
print(frecuencias_relativas)
```

```python
sns.catplot(x="alive",
    hue="class",
    kind="count", # Me permite diferentes
    edgecolor=".6",
    orient = "V",
    data=df_titanic.sort_values("class"));
```

```python
# Tabla de frecuencias modo 2
sns.catplot(x="alive",
            col = "class",
            kind="count",
            edgecolor=".6",
            orient = "V",
            hue = "alive",
            legend= True,
            data=df_titanic);
```

Si queremos frecuencias relativas, primero tenemos que calcularlas y luego ya aplicamos una visualizacin:

```python
#AC

cat_colx = "alive"
cat_coly = "class"
frecuencias_absolutas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts()
frecuencias_relativas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts(normalize=True)
print(frecuencias_absolutas)
print(frecuencias_relativas)
```

```python
frecuencias_relativas["proportion"] *= 100
```

```python
# Pintamos modo1
sns.catplot(x="alive",
            y = "proportion",
            col="class",
            hue = "alive",
            kind="bar", # Ojo, cambio el tipo de grfico (ahora no "cuento" filas)
            edgecolor=".6",
            data=frecuencias_relativas);
```


### Tablas de contingencia y Mapas de Calor

Otra forma numrica de ver las frecuencias de aparicin conjunta de los valores de dos variables categricas es recurrir a las tablas de contingencia:

```python
#AC
tabla_contingencia = pd.crosstab(df_titanic["embark_town"],
    df_titanic["who"],
    margins = False)
tabla_contingencia
```

Comprala con nuestra forma de obtener frecuencias absolutas:

```python
cat_colx = "who"
cat_coly = "embark_town"
frecuencias_absolutas = df_titanic.groupby(cat_coly, as_index = False)[cat_colx].value_counts()
frecuencias_absolutas
```

La primera puede que te resulte ms til a ti visualmente para analizar, la seguda es ms manejable en trminos de programacin.

#### Mapa de calor de la tabla de contigencia:

Si queremos que la tabla de contingencia sea ms "presentable", la convertimos en un mapa de calor:

```python
#AC
#AC
plt.figure(figsize=(5,5))
sns.heatmap(tabla_contingencia,
    vmin = 0,
    vmax = 500,
    cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7),
    square=True,
    linewidths=-1,
    annot=True);
```

Bastante visual,eh? Incluso ms que las barras de frecuencia, si es lo que quiero mostrar... Pero es engorrosa... De usarla cmo lo hara yo:

```python

plt.figure(figsize=(5,5))
sns.heatmap(tabla_contingencia,
            vmin = 10,
            vmax = 200,
            cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7),
            square = True,
            linewidths=.1,
            annot=False);
plt.xlabel("")
plt.ylabel("");
```
