### Ejercicio 1
Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.<funcion>, no usando objetos Axes)

```python
fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,100)
ax.plot(x,np.sin(x));
plt.title(r"Seno")
plt.xticks(np.arange(0, 10.1, 2.5)) 
plt.gca().xaxis.set_major_formatter('{x:.1f}') 


fig = plt.figure()
ax = plt.axes()

x=np.linspace(0,10,100)
ax.plot(x,np.cos(x));
plt.title(r"Coseno")
plt.xticks(np.arange(0, 10.1, 2.5)) 
plt.gca().xaxis.set_major_formatter('{x:.1f}')  

```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos Axes. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)


```python
# Configuraci√≥n inicial
x = np.linspace(0, 10, 500)  # 500 puntos para suavizar curvas

# Crear figura con 3 subplots en columna
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

# Gr√°fico 1: Seno
axs[0].plot(x, np.sin(x), color='red', linewidth=2)
axs[0].set_ylabel('Seno', fontsize=12)
axs[0].set_ylim(-1.1, 1.1)  # L√≠mites del eje y
axs[0].grid(linestyle='--', alpha=0.5)

# Gr√°fico 2: Coseno 
axs[1].plot(x, np.cos(x), color='green', linewidth=2)
axs[1].set_ylabel('Coseno', fontsize=12)
axs[1].set_ylim(-1.1, 1.1)
axs[1].grid(linestyle='--', alpha=0.5)

# Gr√°fico 3: Tangente 
axs[2].plot(x, np.tan(x), color='blue', linewidth=2)
axs[2].set_ylabel('Tangente', fontsize=12)
axs[2].set_ylim(-40, 20)  # Ajuste para valores extremos
axs[2].set_xlabel('x', fontsize=12)
axs[2].grid(linestyle='--', alpha=0.5)

```

### Ejercicio 4: Ajuste de Colores y Estilos de L√≠nea.

Reproduce la figura siguiente. Recuerda que las funciones seno hiperb√≥lico y coseno hiperb√≥lico se obtienen as√≠:

np.sinh, seno hiperb√≥lico;
np.cosh, coseno hiperb√≥lico;

```python
plt.figure(figsize=(18, 2))
# Crear datos
x = np.linspace(0, 10, 10000)
sinh = np.sinh(x)

# Primer gr√°fico: Seno Hiperb√≥lico
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, posici√≥n 1
plt.plot(x, sinh, color='blue', linewidth=2)
plt.title('Seno Hiperb√≥lico', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('sinh(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.figure(figsize=(18, 2))
# Crear datos
x = np.linspace(0, 10, 10000)
cosh = np.cosh(x)

# Segundo gr√°fico: Coseno Hiperb√≥lico
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, posici√≥n 2
plt.plot(x, cosh, color='red', linewidth=2)
plt.title('Coseno Hiperb√≥lico', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('cosh(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosici√≥n (tal y como ves en la figura), emplea subplots_adjust y los par√°metros hspace y wspace.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Datos comunes
x = np.linspace(0, 10, 100)

# Primer subplot: Funci√≥n sinusoidal
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_ylim(-1.1, 1.1)
axs[0, 0].grid(True)

# Segundo subplot: Funci√≥n coseno
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_ylim(-1.1, 1.1)
axs[0, 1].grid(True)

# Tercer subplot: Funci√≥n polin√≥mica
axs[1,0].plot(x, np.tan(x))
axs[1, 0].set_ylim(-45, 25)
axs[1, 0].grid(True)


axs[1, 1].plot(x, np.exp(x/2))  
axs[1, 1].set_ylim(-0.1, 1.6)
axs[1, 1].set_title  # T√≠tulo m√°s descriptivo
axs[1, 1].set_xlabel('x')  # A√±adido etiqueta del eje X
axs[1, 1].set_ylabel('f(x)')  # A√±adido etiqueta del eje Y
axs[1, 1].grid(True, linestyle='--', alpha=0.7)  # Cuadr√≠cula mejorada

# Ajuste de espacios
plt.subplots_adjust(wspace=0.3, hspace=0.4)

plt.show()

# Mostrar figura
plt.show()
```
